# Коллекции
**Коллекция** в Python — программный объект (переменная-контейнер), хранящая набор значений одного или различных типов, позволяющий обращаться к этим значениям, а также применять специальные функции и методы, зависящие от типа коллекции.

Встроенные коллекции в Python:

> Список `list`\
Кортеж `tuple`\
Строка `str`\
Множество `set`\
Неизменяемое множество `frozenset`\
Словарь `dict`

image.png

image.png

**Индексированность** – каждый элемент коллекции имеет свой порядковый номер — индекс. Это позволяет обращаться к элементу по его порядковому индексу, проводить слайсинг («нарезку») — брать часть коллекции выбирая исходя из их индекса. Детально эти вопросы будут рассмотрены в дальнейшем в отдельной статье.

**Уникальность** – каждый элемент коллекции может встречаться в ней только один раз. Это порождает требование неизменности используемых типов данных для каждого элемента, например, таким элементом не может быть список.

**Изменяемость коллекции** — позволяет добавлять в коллекцию новых членов или удалять их после создания коллекции.


# Строка `str`

Встроенная **неизменяемая** коллекция в Python, позволяющая хранить информацию текстового характера. Может состоять из любого количества символов. Поддерживаются различные раскладки и языки, в том числе кириллица.

В Python есть несколько способов для создания строки.

* Первый из них - это одинарные апострофы `'текст'`. Первый апостроф открывает строку, затем следует сам текст, и вторым апострофом вы закрываете строку.
```
print('Моя первая строка')
```
* Второй способ - это кавычки `"текст"`. Также первыми кавычками вы открываете строку, пишете текст и закрываете строку кавычками.
```
print("Моя вторая строка, созданная при помощи кавычек")
```
* Третий способ необходим для создания многострочного текста и также используется для многострочных комментариев.
```
"""
Многострочный
комментарий
"""
text = """
Так можно
сохранить текст 
в переменную
"""
```

Строки можно складывать между собой - операция **конкатенации**:
```
'abc' + 'bcd'
>>> abcbcd
```

Строки можно умножать на какое-то число и тем самым производить **дублирование**:
```
'abc' * 5
>>> abcabcabcabcabc
```

Строки можно **сравнивать** между собой в лексикографическом порядке при помощи стандартных операций сравнения: `==`, `!=`, `>`, `<`, `>=`, `<=`
```
print('abc' > 'bcd')
>>> False
```

При помощи оператора `in` можно проверить **вхождение** подстроки в строку:
```
'a' in 'abc'
>>> True
```

Строка - это **индексируемый** объект, а значит к каждому ее элементу можно обратиться отдельно при помощи индекса. Индексация в Python начинается с 0:
```
s = 'abc'
s[0]
>>> a
```




 




## Срезы

В Python существует мощный инструмент под названием срезы (от англ. *slices*). Он позволяет обращаться сразу к нескольким элементам "по срезу" одной коллекции. Работают срезы для индексируемых коллекций, то есть для строк, списков и кортежей.

Синтаксис конструкции срезов:
```
коллекция[стартовый_индекс:{последний_индекс:{шаг}}]
```
Обратите внимание, что "последний индекс" никогда не включается в вывод. А первый включается всегда.


Рассмотрим несколько примеров применения срезов в ячейке с кодом на примере строки 'Python'. Запустите ячейку ниже:

````python
s = 'Python'
print('Второй символ строки: ', s[2])
print('Последний индекс строки: ', s[-1])
print('Слайс с 2 по 4 символы: ', s[2:4])
print('Перевернутая строка: ', s[::-1])
print('Перевернутый срез с 2 по 4 символы: ', s[4:2:-1])
````

# Циклы

Коллекции состоят из элементов и часто нам необходимо итерироваться по элементам этой коллекции, то есть обходить коллекцию элементов по одному элементу за раз. Сделать это можно при помощи циклов.

В Python существует две разновидности циклов: `for` и `while`.

## Цикл `for`

Цикл `for` позволяет поочередно обойти все элементы итерируемой последовательности или итерируемого объекта. Конструкция имеет следующий синтаксис:
```
for <переменная> in <последовательность>:
    <тело цикла>
```

Например вот таким образом можно обойти все буквы некоторой строки и вывести их на экран:
```
for letter in "text":
    print(letter)
>>> t
>>> e
>>> x
>>> t
```

То есть мы поочередно перебираем все элементы данной коллекции элементов и выводим элементы на экран. **Данная конструкция справедлива для всех коллекций Python!**

Также конструкцию цикла `for` можно применять совместно с функцией `range()`. 
`range()` - это особая функция-генератор, способная генерировать арифметические прогрессии. Синтаксис у нее такой: 
```
range(конечный_элемент)
range(начальный_элемент, конечный_элемент)
range(начальный_элемент, конечный_элемент, шаг)
```

Однако, если вы просто запустите в ячейке функцию, например, `range(10)`, то вы не получите следующий вывод:

```
range(10)
>>> range(0, 10)
```
Все дело в том, что `range` - это особый вид функции, называемой **генератором**. Подробнее они будут разбираться на продвинутом курсе. А если вам нужно получить список всех элементов арифметической прогрессии, то можно обернуть эту конструкцию в функцию `list()`:
```
list(range(5))
>>> [0, 1, 2, 3, 4]

list(range(4, 7))
>>> [4, 5, 6]

list(range(5, -1, -1))
>>> [5, 4, 3, 2, 1, 0]
```

**Важно!**\
Последний элемент не входит в генерируемую коллекцию.








## Цикл `while`

Цикл `while` это наиболее часто встречаемая конструкция в алгоритмах, так как это более гибкая конструкция в сравнении с циклом `for`. Синтаксис данной конструкции такой:

```
while <условие>:
  <тело цикла>
```

Если необходимо итерироваться по коллекции, то необходимо создать переменную, которая будет отвечать за итерации. Например вот так:

```
i = 0
while i < 3:
  print(i)
  i += 1

>>> 0
>>> 1
>>> 2
```

Часто эту конструкцию используют для создания бесконечных циклов. Например:

```
while True:
  <тело цикла>
```

Однако, для того, чтобы выйти из такого цикла, необходимо предусмотреть оператор `break`. Он будет прерывать выполнение бесконечного цикла. Например:

```
i = 0
while True:
  if i == 4:
    break
  print(i)
  i += 1

>>> 0
>>> 1
>>> 2
```

Оператор `break` также применим к циклу `for`.

# Список `list`

Список в Python - это упорядоченная индексируемая и **изменяемая** коллекция элементов. Является динамическим массивом, элементы могут быть разных типов данных.

Список можно создать при помощи `[]` или функции `list()`. Рассмотрим пример создания простого списка:
```
my_list = [23, 42, 98, 23]
```
Элементы списка не являются уникальными и могут повторяться. Поскольку список - это индексируемая коллекция, то к каждому ее элементу можно обратиться по индексу. Напомню, что индексация начинается с нуля. Например:
```
my_list = [23, 42, 98, 23]
my_list[0]
>>> 23
```
Поскольку списки являются изменяемой коллекция, то также можно изменить элемент списка по его индексу:

```
my_list = [23, 42, 98, 23]
my_list[0] = 'измененный элемент'
my_list
>>> ['измененный элемент', 42, 98, 23]
```

При работе со списками необходимо быть осторожным, так как в отличиие от **неизменяемых** объектов (int, float, bool, str, tuple, frozenset), изменяемые объекты (list, set, dict) будут изменеться по любому указателю на них. А поскольку в Python любая переменная это и есть указатель, то изменяемые объекты будут меняться по любой из этих ссылок. Попробуйте код, представленный ниже самостоятельно в ячейке колаба:
```
a = [1, 2, 3]
b = a
print(b)
b[0] = 'Изменили 0 элемент в объекте под переменной "b"'
print(b)
print(a) # Однако, это изменение повлекло за собой изменение в объекте под переменной "а"
```

Списки можно **конкатенировать** между собой:
```
[1, 2, 3] + ["a", 'b', "dwf"]
>>> [1, 2, 3, 'a', 'b', 'dwf']
```

Списки можно **умножать** на целые числа:
```
[1, 2, 3] * 3
>>> [1, 2, 3, 1, 2, 3, 1, 2, 3]
```

Можно найти **длину** списка:
```
len([1, 2, 3])
>>> 3
```

Список можно отсортировать при помощи функции **sorted()**:
```
sorted([3, 5, 1, 2])
>>> [1, 2, 3, 5]
```

Также для списков аналогичным образом работают **срезы**:
```
my_list = [1,2,3]
my_list[:2] 
>>> [1, 2]
```

При помощи стандартных функций **min** и **max** можно находить наибольшее и наименьшее значений списка:
```
my_list = [1,2,3]
print(min(my_list), max(my_list))
>>> 1 3
```

Также можно найти сумму элементов списка, состоящих из чисел:
```
my_list = [1,2,3]
print(sum(my_list))
>>> 6
```

# Кортеж `tuple`

Кортеж (tuple) – **неизменяемая** упорядоченная последовательность.  Кортеж очень напоминает список, но только кортеж является неизменяемым типом данных.

Создать кортеж можно двумя способами:

```
my_tuple = (1, 2, 3) # При помощи круглых скобок
my_tuple = tuple([1, 2, 3]) # При помощи функции tuple, в которую можно передать список элементов
```

После создания кортежа, его элементы нельзя изменять:

```
my_tuple = (1, 2, 3)
my_tuple[0] = 'Новый элемент'
>>> TypeError: 'tuple' object does not support item assignment
```

Кортежи часто применяетсю в алгоритмах, если мы боимся случайно изменить какую-то хранимую информацию, например, в графовых алгоритмах поиска.

Большинство методов, которые работают со списками также будут работать и с кортежами, за исключением тех, кто напрямую меняют списки.

Примеры работы с кортежами представлены ниже:
```
my_tuple = tuple()  # --> пустой кортеж
my_tuple = ()  # --> пустой кортеж
my_tuple = (1, 2, 3, 4, 5)  # --> кортеж (1, 2, 3, 4, 5)
# индексация и слайсинг
my_tuple[1]  # --> 2
my_tuple[2:4]  # --> (3, 4)

# Всегда можем перейти от кортежа к списку и наоборот
list_variable = list(my_tuple)
tuple_variable = tuple(list_variable)
```




# Множество `set`

Множество - изменная **неупорядоченная** коллекция **уникальных** элементов.
Удобно в ситуациях, когда нужно обработать некий поток данных из которого нужно взять только уникальные значения. Или когда нужно произвести операции выборки определенного подмножества из нескольких данных. Например, взять только те значения, которые есть в наборе `a` и наборе `b`.

```
# Варианты создания
my_set = set()  # --> пустое множество
my_set2 = {1, 2, 3}  # --> множество из элементов {1, 2, 3}
my_set3 = {1, 2, 2, 3}  # --> множество из элементов {1, 2, 3}
my_set3.add(4)  # --> добавляет 4 к множеству
my_set3.pop()  # --> достает из множества случайный элемент
```

Популярные методы работы с множествами:
* Мощность множества `len(<множество>)`
* Добавление элемента в множество `<множество>.add(element)`
* Удаление элемента по значению `<множество>.discard(element)`
* Удаление произвольного элемента и передает удаленное значение на экран `<множество>.pop()`

Также над множеством можно осуществлять математические операции, они представлены на картинке:

image019-4.png

```
# Математические операции над множествами
A = {1, 3, 5, 7}
B = {3, 7, 6, 4, 9, 10, 1}
print('Пересечение множеств: ', A & B)
print('Объединение множеств: ', A | B)
print('Симметричная разница: ', A ^ B)
print('Разница множеств A-B:', A - B)
print('Разница множеств B-A:', B - A)
```

# Словарь `dict`

Словарь `dict` (его еще называют ассоциативным списком) – **неупорядоченная** совокупность произвольных объектов с **доступом по ключу**. Словари это очень важный тип данных, вы будете с ним очень часто работать, возможно так же часто как и со списками. Ключ играет роль индекса элементов.

* Хранилище данных типа ключ-значение. Близко к классической структуре данных HashTable.
* Например, может выступать как мини база данных. Размер и уровень вложенности не ограничен, хранить можно объекты любого типа.
* Ключи уникальны в пределах одного словаря.
В качестве ключей могут выступать любые неизменяемые (immutable) встроенные типы.

Варианты создания словаря представлены ниже:
```
my_dict = dict()  # --> пустой словарь {}
my_dict2 = {}  # --> пустой словарь {}
ages = {
   "Dmitry": 26,
   "Alex": 34
}  # --> словарь с ключами "Dmitry" и "Alex"

del ages["Dmitry"]  # --> удалит ключ Dmitry из словаря
```

Популярные методы работы можно найти вот здесь:
```
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
print('Обращение по ключу "a": ', my_dict['a'])
my_dict['a'] = 't'
print('Поменять значение в словаре по ключу "a": ', my_dict)
my_dict['new_key'] = 'new_value'
print('Добавление новой пары ключ-значение: ', my_dict)
print('Проверить если ключ "b" есть в словаре: ', 'b' in my_dict)
print('Все ключи словаря, преобразованные в список: ', my_dict.keys())
print('Все значения словаря, преобразованные в список: ', my_dict.values())
```

### Задача 23. Два списка

На вход даны два списка. Выведите элемент первого списка, если его нет во втором списке.

**Ввод**
```
89 98 64 90 68 26 51 21 91
18 75 15 17 33 80 63 98 100
```

**Вывод**
```
89 64 90 68 26 51 21 91 
```

---

**Ввод**
```
8 7 9 5 1 7 6 8
3 9 9 8 5 5 3 4 5
```

**Вывод**
```
7 1 7 6 
```


````python
# Ваше решение
a = list(map(int, input().split()))
b = list(map(int, input().split()))

c = []
for i in a:
  if not(i in b):
    c.append(i)
print(c)
````
