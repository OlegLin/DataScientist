## Python
`Python` (или *пайтон*, а также *питон*) - *интерпретируемый* объектно-ориентированный язык программирования с *динамической типизацией*.

**Интерпретируемость** означает, что код вашей программы выполняется построчно. Чтобы запустить вашу программу, компьютер будет выполнять ее строчка-за-строчкой. Если в конце вашего кода вы допустили ошибку, то вы не узнаете о ней, пока компьютер не дойдет до нее.

Другая основная группа языков программирования называется *компилируемыми*. Примером компилируемых языков являются языки C/C++. Чтобы выполнять программу на подобных языках программирования, компьютер сначала собирает её, подключая необходимые библиотеки, переводит в вид, который будет понятен для него, проверяет на наличие ошибок, а уже затем позволяет выполнить такую программу.

**Динамическая типизация** означает, что программисту не нужно указывать тип переменной. Компьютер сам определит какой тип переменной был задан. Например, вот так можно задать переменную с целочисленным типом:

```
a = 112
```

А потом если вы передумали и решили использовать эту переменную для строк или символов, то никакой проблемы:

```
a = 'мой текст'
```
С другими языками программирования со статической типизацией, например С/С++, так работать нельзя. Если вы объявили, что переменная хранит числа, то компьютер всегда будет ожидать в этой переменной численный тип.

Код на языке Python можно выполнять в двух вариантах:

Однако, чтобы компьютер сумел запустить программу, ему необходим помощник - [интепретатор](https://www.python.org/downloads/) языка. Он будет выполнять роль переводчика текста, который понятен программисту (синтаксис языка) в текст, понятный компьютеру (бинарный файл, состоящий из 0 и 1).

`print(54 > 55)` В этом случае сравниваются 2 числа и результат сравнения — объект логического типа. Конкретно в этом случае результатом исполнения будет False.

В `Python` существует очень много различных типов. В таблице представлено большинство из них.

Сейчас давайте разберем 2 типа объектов: целые числа(int, от слова "integer") и числа с плавающей точкой(float, число с дробной частью).

> *Целое число - число, которое не содержит дробную часть. К ним относятся все положительные, все отрицательные, а также 0. Например: 7, 6, 5, 2, 100, -5, -99.*

> *Вещественные числа - числа, которые содержат дробную часть. Целая часть в Python отделяется от дробной знаком "точки". Если вы вместо точки поставите запятую, то Python воспримет это как перечисление двух целых чисел.
Пример: 6.3, 7.5, -9.6.*

## Операции над числами

В программировании, как и в математике, над числами можно выполнять различные операции. Многие из них вы знаете из математики. Все основные математические операции представлены в таблице:

image.png

Давайте рассмотрим некоторые из них на практике. Для этого просто запускайте каждый последующий пример в колабе.

При сложении двух **целых** чисел результатом будет также целое число:





````python
3 + 2
````

Если сложить **вещественное** число с целом, то результатом будет уже вещественное число - оба числа будут преобразованы к типу **float**:



````python
1.3 + 7
````

Конечно, если складывать два вещественных числа, то получится также **вещественное**:

````python
4.5 + 5.9
````

Оперрация **вычитания** обозначается через дефис. Логика исполнения такая же, как и при сложении:

````python
8.9 - 6.9
````

````python
6 - 3
````

````python
6.9 - 5
````

**Умножение** обозначается звездочкой (астерикс).

````python
2 * 4
````

````python
5.3 * 2
````

Следующая операция - **деление**. Здесь есть 1 нюанс - при делении целого числа  вы ВСЕГДА  будете получать вещественное:

````python
2 / 3
````

````python
2.1 / 1
````

Полезная операция - **возведение в степень**. Сначала вы указываете кого возводите в степень, затем уже саму степень (помните, что в общем случае операция возведения в степени - *неассоциативна*, то есть $(a^b)^c \neq a^{(b^c)}$:

````python
2 ** 3
````

Следующая операция - **деление нацело**. Она отличается от обычного деления тем, что ответ всегда будет целого типа. И результат получается исходя из ответа на вопрос "Сколько раз второе число помещается в первое?".

````python
14 // 3
````

## Конструкция `if-elif-else`

Логические конструкции, рассмотренные ранее очень часто применяются в алгоритмах ветвления, которые в Python можно построить при помощи конструкции `if`.
Существует три базовых варианта написания данной конструкции:
1. `if` - когда просто хотим проверить одно выражение на истинность
Пример:
```
a = 10
if a == 10:
    print('Число а равно 10!')
```
Если условие не выполнится, то мы просто пропустим вложенные инструкции.

2. `if-else` - когда есть альтернативное развитие событий в случае невыполнения программы
Пример:
```
a = 11
if a == 10:
    print('Число а равно 10!')
else:
    print('Число а не равно 10!')
```

3. `if-elif-else` - когда существует несколько альтернативных вариантов. вложенность elif (от англ. 'else if' - 'иначе если') может сколь угодно длинной.
Пример:
```
a = 11
if a == 10:
    print('Число а равно 10!')
elif a == 11:
    print('Число а равно 11!')
else:
    print('Число а не равно 10, и не равно 11!')
```
